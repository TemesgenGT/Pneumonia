## Subset the dataset into training, validation and test classes

```python
train_gen = ImageDataGenerator(rescale = 1./255.,
                                   zoom_range = 0.2)
val_gen = ImageDataGenerator(rescale = 1./255.,)
test_gen = ImageDataGenerator(rescale = 1./255.,)


train_ds = train_gen.flow_from_directory(train_dir, 
                                         batch_size=32, 
                                         class_mode='binary', target_size = (220, 220))
val_ds = val_gen.flow_from_directory(val_dir,
                                     batch_size=8, 
                                     class_mode = 'binary', 
                                     target_size=(220, 220))
test_ds = test_gen.flow_from_directory(test_dir,
                                       shuffle=False, 
                                       batch_size=16, 
                                       class_mode = 'binary', 
                                       target_size=(220, 220))
```

## Convolution Neural Network Deep Learing Modeling

```python
# input shape fed to first layer of the cnn
input_shape = train_ds.image_shape 
# using Sequential method connect the layers 
model = Sequential()
# apply Conv2D and relu activation to filter pixes with 0 and above values.
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, padding='same'))
# Filter the data further by taking the maximum value of pixles from 2 by 2 blocks
model.add(MaxPooling2D(pool_size=(2, 2)))
# repeat the same steps until the desired pexiles remain.
model.add(Conv2D(32, (3, 3),  activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu', padding="same"))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu', padding="same"))
model.add(MaxPooling2D(pool_size=(2, 2)))
#Flatten the networked multilayer values into a single layer
model.add(Flatten())

model.add(Dense(64, activation="relu"))

model.add(Dropout(0.5))

model.add(Dense(1))
# Sigmoid function classify the result either the image is Normal or has Pnuemonia symptom.
model.add(Activation('sigmoid'))
```

## Compile the Model

```python
# compile all the relevant metrics result for the model
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
# Create Input Shape
input_shape = (None,220,220,3)
```

```python
# build the model and pass the input shape
model.build(input_shape)
```

```python
# Inspect the summary
model.summary()
```

## Fit and train the model

```python
# Train the model, fitting both the training and test datasets, number of epochs etc.
training = model.fit(
        train_ds,
       validation_data= val_ds,
        steps_per_epoch=60,
        epochs=20)
```

## Evalueate the Model

```python
# check model performance 
model.evaluate(test_ds)
```

```python
# Plot the accuracy from training and validation data
history = training.history
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.plot(history['accuracy'], color='b')
plt.plot(history['val_accuracy'], color = 'orange')

plt.title('Accuracy')
plt.legend(['accuracy', 'val_accuracy'], loc='best')
# plot the loss from the training and validation data
plt.subplot(1,2,2)
plt.plot( history['loss'])
plt.plot( history['val_loss'])

plt.title("Loss")
plt.legend(['Loss'], loc='best')
plt.show()
```

## Predict the model using the unseen data(test data)

```python
# predict the test data
pred = model.predict(test_ds)
```

```python
# group the predicted  as 1 if prob is > 0.5 else 0
predicted = []
for  prob in pred:
    if prob  < 0.5:
         predicted.append(0)
    else:
        predicted.append(1)
```

Assign the expected value for y_true

```python
# The true predicted value is the same of the test dataset values 
y_true = test_ds.classes
```

```python
# Check the dimension to make sure all the test dataset are returned 
len(y_true)
```

## Confusion Matrix 

```python
# Apply confusion metrix to compare how much of the predicted identified correctly
cm =confusion_matrix(y_true, predicted)
```

```python
# Summary of metrics on model performance 
print(classification_report(y_true, predicted))
```

The most relevant aspect of this modeling is how much of the people who are labeled as pneumonia are accurately classified that defined by recall or sensitivity parameter. Based on the recall parameter, the model returned 98% of the patients identified as having pneumonia. The accuracy score from the imbalance date is not a reliable outcome, therefore, choosing F-score is appropriate to evaluate model performance.  F score selection is also associated with the target outcome of the model. F score is the harmonic mean of precision and recall. When F score is One is a harmonic mean of precision and recall parameters, and F score greater than One is model improve recall outcome at the cost of precision,  and the reverse is true for F score less than One.  This project targeted better performance in recall than precision hence, getting the optimal threshold is the key to model performance.  

```python
# Use data augmentation to propagate thw number of data from the existing dataset using Random Flip and Random rotation
data_augmentation = tf.keras.Sequential([
tf.keras.layers.experimental.preprocessing.RandomFlip('horizontal'),
tf.keras.layers.experimental.preprocessing.RandomRotation(0.2)])
```

### Incooperate data_augmentation techinique

```python
# input shape fed to first layer of the cnn
input_shape = train_ds.image_shape 
# using Sequential method connect the layers 
model_1 = Sequential()
# apply Conv2D and relu activation to filter pixes with 0 and above values.
model_1.add(data_augmentation)
model_1.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, padding='same'))
# Filter the data further by taking the maximum value of pixles from 2 by 2 blocks

model_1.add(MaxPooling2D(pool_size=(2, 2)))
# repeat the same steps until the desired pexiles remain.

model_1.add(Conv2D(32, (3, 3),  activation='relu', padding='same'))

model_1.add(MaxPooling2D(pool_size=(2, 2)))

model_1.add(Conv2D(64, (3, 3), activation='relu', padding="same"))
model_1.add(MaxPooling2D(pool_size=(2, 2)))

model_1.add(Conv2D(64, (3, 3), activation='relu', padding="same"))
model_1.add(MaxPooling2D(pool_size=(2, 2)))
#Flatten the networked multilayer values into a single layer
model_1.add(Flatten())

model_1.add(Dense(64, activation="relu"))

model_1.add(Dropout(0.5))

model_1.add(Dense(1))
# Sigmoid function classify the result either the image is Normal or has Pnuemonia symptom.
model_1.add(Activation('sigmoid'))
```

```python
# compile all the relevant metrics result from the model
model_1.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
# Create Input Shape
input_shape = (None,220,220,3)
```

```python
# build the model and pass the input shape
model_1.build(input_shape)
```

```python
# Inspect the summary
model_1.summary()
```

```python
# Fit and Training the dataset using train_ds, val_ds
training = model_1.fit(
        train_ds,
        validation_data= val_ds,
        steps_per_epoch = 60,
        epochs=20)
```

```python
# The accuracy of the model after data augmentation techenique
model_1.evaluate(test_ds)
```

Summary:
The model performance is found better without applying data augmentation. The training imbalance data
reveals the data is enough to train the model thus no overfitting problem is detected. On the contrary,  implementing the data augmentation technique exacerbate the data imbalance problem, as a result, the model performance declined compared to the model without data augmentation.
Generally, imbalance data, data insufficiency, and data augmentation technique seek meticulous evaluation.
